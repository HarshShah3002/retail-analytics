```python
# ── RFM ANALYSIS ─────────────────────────────────────────────
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
from datetime import timedelta

# Connect and load data
engine = create_engine('mysql+pymysql://root:@localhost/retail_analytics')
df = pd.read_sql('SELECT * FROM transactions', engine)

# We need a reference date - we'll use 1 day after the last invoice date
reference_date = df['invoice_date'].max() + timedelta(days=1)
print(f"Reference date: {reference_date}")
print(f"Last transaction date: {df['invoice_date'].max()}")
```

```python
# ── STEP 1: Calculate RFM Metrics ────────────────────────────

rfm = df.groupby('customer_id').agg(
    recency=('invoice_date', lambda x: (reference_date - x.max()).days),
    frequency=('invoice', 'nunique'),
    monetary=('total_amount', 'sum')
).reset_index()

# Round monetary to 2 decimal places
rfm['monetary'] = rfm['monetary'].round(2)

print(f"Total customers analysed: {len(rfm)}")
print("\nRFM Sample:")
print(rfm.head(10))
print("\nRFM Statistics:")
print(rfm[['recency', 'frequency', 'monetary']].describe())
```

```python
# ── STEP 2: Score each customer 1-5 on R, F, M ───────────────

# For Recency: lower days = better = higher score (so we reverse it)
rfm['r_score'] = pd.qcut(rfm['recency'], q=5, labels=[5, 4, 3, 2, 1])

# For Frequency: higher = better = higher score
rfm['f_score'] = pd.qcut(rfm['frequency'].rank(method='first'), q=5, labels=[1, 2, 3, 4, 5])

# For Monetary: higher = better = higher score
rfm['m_score'] = pd.qcut(rfm['monetary'], q=5, labels=[1, 2, 3, 4, 5])

# Convert to integers
rfm['r_score'] = rfm['r_score'].astype(int)
rfm['f_score'] = rfm['f_score'].astype(int)
rfm['m_score'] = rfm['m_score'].astype(int)

# Combine into one RFM score
rfm['rfm_score'] = rfm['r_score'].astype(str) + rfm['f_score'].astype(str) + rfm['m_score'].astype(str)

print("RFM Scores Sample:")
print(rfm[['customer_id', 'recency', 'frequency', 'monetary', 'r_score', 'f_score', 'm_score', 'rfm_score']].head(10))
```

```python
# ── STEP 3: Assign Segment Labels ────────────────────────────

def assign_segment(row):
    r = row['r_score']
    f = row['f_score']
    m = row['m_score']
    
    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'
    elif r >= 3 and f >= 3 and m >= 3:
        return 'Loyal Customers'
    elif r >= 4 and f <= 2:
        return 'Potential Loyalists'
    elif r >= 3 and f <= 2:
        return 'Recent Customers'
    elif r <= 2 and f >= 4 and m >= 4:
        return 'At Risk'
    elif r <= 2 and f >= 3 and m >= 3:
        return 'Needs Attention'
    elif r == 1 and f <= 2 and m <= 2:
        return 'Lost'
    else:
        return 'Others'

rfm['segment'] = rfm.apply(assign_segment, axis=1)

# Show segment counts
segment_counts = rfm['segment'].value_counts()
print("Customer Segments:")
print(segment_counts)
print(f"\nTotal customers: {len(rfm)}")
```

```python
# ── STEP 4: Visualise Segments ───────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Chart 1 - Customer count per segment
colors = ['#2ecc71', '#27ae60', '#e74c3c', '#95a5a6', 
          '#3498db', '#f39c12', '#1abc9c', '#e67e22']

segment_counts.plot(kind='bar', ax=axes[0], color=colors, edgecolor='white')
axes[0].set_title('Number of Customers per Segment', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Segment')
axes[0].set_ylabel('Number of Customers')
axes[0].tick_params(axis='x', rotation=45)

# Add value labels on bars
for i, v in enumerate(segment_counts.values):
    axes[0].text(i, v + 10, str(v), ha='center', fontsize=10)

# Chart 2 - Revenue per segment
segment_revenue = rfm.groupby('segment')['monetary'].sum().sort_values(ascending=False)

segment_revenue.plot(kind='bar', ax=axes[1], color=colors, edgecolor='white')
axes[1].set_title('Total Revenue per Segment', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Segment')
axes[1].set_ylabel('Total Revenue (£)')
axes[1].tick_params(axis='x', rotation=45)

# Add value labels on bars
for i, v in enumerate(segment_revenue.values):
    axes[1].text(i, v + 1000, f'£{v:,.0f}', ha='center', fontsize=9)

plt.suptitle('RFM Customer Segmentation Analysis', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('notebooks/chart9_rfm_segments.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved!")
```

```python
# ── STEP 5: Save RFM results to MySQL ────────────────────────
engine = create_engine('mysql+pymysql://root:@localhost/retail_analytics')

rfm.to_sql('rfm_segments', con=engine, if_exists='replace', index=False)
print(f"RFM table saved to MySQL with {len(rfm)} rows!")

# Preview final RFM table
print("\nFinal RFM Table Sample:")
print(rfm[['customer_id', 'recency', 'frequency', 'monetary', 
           'r_score', 'f_score', 'm_score', 'rfm_score', 'segment']].head(10))
```

```python
# ── PHASE 5: ML SALES FORECASTING ────────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Load data from MySQL
engine = create_engine('mysql+pymysql://root:@localhost/retail_analytics')
df = pd.read_sql('SELECT * FROM transactions', engine)

# ── Create monthly revenue dataset ───────────────────────────
df['invoice_date'] = pd.to_datetime(df['invoice_date'])
df['year_month'] = df['invoice_date'].dt.to_period('M')

monthly = df.groupby('year_month').agg(
    revenue=('total_amount', 'sum'),
    orders=('invoice', 'nunique'),
    customers=('customer_id', 'nunique')
).reset_index()

monthly['year_month'] = monthly['year_month'].astype(str)
monthly['revenue'] = monthly['revenue'].round(2)

print(f"Monthly data shape: {monthly.shape}")
print("\nMonthly data:")
print(monthly)
```

```python
# ── STEP 2: Feature Engineering ──────────────────────────────
# Feature engineering = creating useful inputs for our ML model
# We can't just feed dates into a model - we need to convert 
# them into numbers that capture patterns

# Remove last row - incomplete month
monthly = monthly[:-1].copy()

# Convert year_month to datetime for feature extraction
monthly['date'] = pd.to_datetime(monthly['year_month'])

# Time based features
monthly['month'] = monthly['date'].dt.month          # 1-12 (captures seasonality)
monthly['year'] = monthly['date'].dt.year            # 2009, 2010, 2011
monthly['month_number'] = range(1, len(monthly)+1)   # 1,2,3... (captures trend)

# Lag features - previous months revenue
# This is powerful - last month's revenue predicts this month's!
monthly['lag_1'] = monthly['revenue'].shift(1)   # Last month
monthly['lag_2'] = monthly['revenue'].shift(2)   # 2 months ago
monthly['lag_3'] = monthly['revenue'].shift(3)   # 3 months ago

# Rolling average - smooths out noise
monthly['rolling_3'] = monthly['revenue'].shift(1).rolling(window=3).mean()

# Season feature (Q4 = Christmas peak)
monthly['is_q4'] = monthly['month'].isin([10, 11, 12]).astype(int)

# Drop rows with NaN from lag features
monthly = monthly.dropna()

print(f"Training data shape: {monthly.shape}")
print("\nFeature sample:")
print(monthly[['year_month', 'revenue', 'month', 'month_number', 
               'lag_1', 'lag_2', 'rolling_3', 'is_q4']].head(10))
```

```python
# ── STEP 3: Train the ML Model ───────────────────────────────

# Define features (inputs) and target (what we want to predict)
features = ['month', 'month_number', 'lag_1', 'lag_2', 'lag_3', 'rolling_3', 'is_q4']
target = 'revenue'

X = monthly[features]
y = monthly[target]

# Split into training and testing sets
# We use the last 6 months as test data (time series split)
# We don't use random split because order matters in time series!
X_train = X[:-6]
X_test = X[-6:]
y_train = y[:-6]
y_test = y[-6:]

print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")
print(f"\nTest period: {monthly['year_month'].iloc[-6]} to {monthly['year_month'].iloc[-1]}")

# Train Random Forest model
model = RandomForestRegressor(
    n_estimators=200,    # 200 decision trees
    max_depth=5,         # Limit tree depth to avoid overfitting
    random_state=42      # For reproducibility
)

model.fit(X_train, y_train)
print("\nModel trained successfully!")

# Make predictions on test set
y_pred = model.predict(X_test)

# Evaluate model performance
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"\nModel Performance:")
print(f"MAE  (Mean Absolute Error):  £{mae:,.0f}")
print(f"RMSE (Root Mean Sq Error):   £{rmse:,.0f}")
print(f"R²   (Accuracy Score):       {r2:.3f}")
```

```python
# ── STEP 4: Visualise Actual vs Predicted ────────────────────

test_months = monthly['year_month'].iloc[-6:].values

plt.figure(figsize=(14, 6))

# Plot full actual revenue line
plt.plot(monthly['year_month'], monthly['revenue'], 
         color='steelblue', linewidth=2.5, marker='o', label='Actual Revenue')

# Plot predicted values on test period
plt.plot(test_months, y_pred, 
         color='coral', linewidth=2.5, marker='s', 
         linestyle='--', label='Predicted Revenue')

# Shade the test period
plt.axvspan(test_months[0], test_months[-1], alpha=0.1, color='coral', label='Test Period')

# Add value labels on predictions
for i, (month, pred) in enumerate(zip(test_months, y_pred)):
    plt.text(month, pred + 20000, f'£{pred:,.0f}', 
             ha='center', fontsize=8, color='coral')

plt.title('Actual vs Predicted Monthly Revenue', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Month')
plt.ylabel('Revenue (£)')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.savefig('notebooks/chart10_forecast.png', dpi=150)
plt.show()
print("Chart saved!")
```

```python
# ── STEP 5: Feature Importance ───────────────────────────────
# Feature importance tells us which inputs the model relied on most

importance = pd.DataFrame({
    'feature': features,
    'importance': model.feature_importances_
}).sort_values('importance', ascending=False)

plt.figure(figsize=(10, 6))
bars = sns.barplot(x='importance', y='feature', data=importance,
                   hue='feature', palette='Reds_r', legend=False)

# Add value labels
for i, v in enumerate(importance['importance'].values):
    bars.text(v, i, f' {v:.3f}', va='center', fontsize=10)

plt.title('Feature Importance — What drives revenue predictions?', 
          fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.tight_layout()
plt.savefig('notebooks/chart11_feature_importance.png', dpi=150)
plt.show()

print("\nFeature Importance:")
print(importance)
```

```python
# ── EXTRA 1: COHORT ANALYSIS ──────────────────────────────────

# Step 1: Find each customer's first purchase month (their cohort)
df['invoice_date'] = pd.to_datetime(df['invoice_date'])
df['order_month'] = df['invoice_date'].dt.to_period('M')

cohort_df = df.groupby('customer_id')['order_month'].min().reset_index()
cohort_df.columns = ['customer_id', 'cohort_month']

# Step 2: Merge cohort month back to main dataframe
df = df.merge(cohort_df, on='customer_id')

# Step 3: Calculate how many months after first purchase each order was made
df['cohort_index'] = (df['order_month'] - df['cohort_month']).apply(lambda x: x.n)

# Step 4: Count unique customers per cohort and cohort index
cohort_data = df.groupby(['cohort_month', 'cohort_index'])['customer_id'].nunique().reset_index()
cohort_data.columns = ['cohort_month', 'cohort_index', 'customers']

# Step 5: Create pivot table
cohort_pivot = cohort_data.pivot_table(index='cohort_month', 
                                        columns='cohort_index', 
                                        values='customers')

# Step 6: Convert to retention percentages
cohort_size = cohort_pivot.iloc[:, 0]
retention = cohort_pivot.divide(cohort_size, axis=0).round(3) * 100

print("Cohort Retention Table (first 6 months):")
print(retention.iloc[:, :6].to_string())
```

```python
# ── COHORT RETENTION HEATMAP ─────────────────────────────────

plt.figure(figsize=(18, 10))

sns.heatmap(retention.iloc[:, :12],  # Show first 12 months
            annot=True,               # Show numbers in cells
            fmt='.0f',                # No decimals
            cmap='YlOrRd_r',          # Darker = better retention
            linewidths=0.5,
            vmin=0, vmax=50,          # Fix scale 0-50%
            cbar_kws={'label': 'Retention %'})

plt.title('Customer Cohort Retention Analysis\n(% of customers returning each month)', 
          fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Months Since First Purchase')
plt.ylabel('Cohort Month')
plt.tight_layout()
plt.savefig('notebooks/chart12_cohort_retention.png', dpi=150)
plt.show()
print("Chart saved!")
```

```python
# ── EXTRA 2: BASKET ANALYSIS ─────────────────────────────────
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Step 1: Prepare basket data
# Group products by invoice - each invoice = one shopping basket
print("Preparing basket data...")

# Focus on UK only - largest market, faster processing
uk_df = df[df['country'] == 'United Kingdom']

# Create list of products per invoice
basket = uk_df.groupby('invoice')['description'].apply(list).reset_index()
basket.columns = ['invoice', 'items']

print(f"Total baskets (invoices): {len(basket)}")
print(f"\nSample basket:")
print(basket['items'].iloc[0][:5])
```

```python
# Step 2: Encode the basket data
print("Encoding basket data...")

basket_sample = basket['items'].head(5000).tolist()

te = TransactionEncoder()
te_array = te.fit_transform(basket_sample)
basket_encoded = pd.DataFrame(te_array, columns=te.columns_)

print(f"Basket matrix shape: {basket_encoded.shape}")
print("Encoding complete!")

# Step 3: Run Apriori algorithm
print("\nRunning Apriori algorithm...")
frequent_itemsets = apriori(basket_encoded, 
                             min_support=0.02,
                             use_colnames=True,
                             max_len=2)

print(f"Frequent itemsets found: {len(frequent_itemsets)}")

# Step 4: Generate association rules
rules = association_rules(frequent_itemsets, 
                          metric='lift',
                          min_threshold=1.5,
                          num_itemsets=len(frequent_itemsets))

rules = rules.sort_values('lift', ascending=False)

print(f"Association rules found: {len(rules)}")
print("\nTop 10 Product Pairs:")
top_rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10)
print(top_rules.to_string())
```

```python
# ── VISUALISE BASKET ANALYSIS ────────────────────────────────

# Clean up product names for display
def clean_name(itemset):
    name = list(itemset)[0]
    return name[:25] + '...' if len(name) > 25 else name

top_20 = rules.head(20).copy()
top_20['product_a'] = top_20['antecedents'].apply(clean_name)
top_20['product_b'] = top_20['consequents'].apply(clean_name)
top_20['pair'] = top_20['product_a'] + ' → ' + top_20['product_b']

fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Chart 1 - Top pairs by lift
sns.barplot(x='lift', y='pair', data=top_20.head(10),
            hue='pair', palette='Reds_r', legend=False, ax=axes[0])
axes[0].set_title('Top 10 Product Pairs by Lift', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Lift Score')
axes[0].set_ylabel('Product Pair')

# Chart 2 - Confidence vs Support scatter
scatter = axes[1].scatter(top_20['support'], 
                          top_20['confidence'],
                          c=top_20['lift'], 
                          cmap='YlOrRd',
                          s=top_20['lift'] * 10,
                          alpha=0.7)

plt.colorbar(scatter, ax=axes[1], label='Lift')
axes[1].set_title('Support vs Confidence\n(bubble size = lift)', 
                   fontsize=13, fontweight='bold')
axes[1].set_xlabel('Support')
axes[1].set_ylabel('Confidence')

plt.suptitle('Market Basket Analysis — Product Associations', 
             fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('notebooks/chart13_basket_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved!")
```

```python
# ── EXTRA 3: FORECAST NEXT 6 MONTHS ──────────────────────────

# Step 1: Rebuild model on ALL available data (not just training set)
# Now that we've evaluated it, we use all data for best predictions
features = ['month', 'month_number', 'lag_1', 'lag_2', 'lag_3', 'rolling_3', 'is_q4']

X_all = monthly[features]
y_all = monthly['revenue']

model_full = RandomForestRegressor(
    n_estimators=200,
    max_depth=5,
    random_state=42
)
model_full.fit(X_all, y_all)
print("Model retrained on full dataset!")

# Step 2: Build next 6 months feature rows
last_row = monthly.iloc[-1]
last_month_num = monthly['month_number'].max()
last_revenues = monthly['revenue'].tolist()

forecast_months = []
forecast_revenue = list(last_revenues)  

future_dates = pd.date_range(start='2012-01-01', periods=6, freq='MS')

for i, date in enumerate(future_dates):
    month_num = last_month_num + i + 1
    month = date.month
    
    lag_1 = forecast_revenue[-1]
    lag_2 = forecast_revenue[-2]
    lag_3 = forecast_revenue[-3]
    rolling_3 = np.mean(forecast_revenue[-3:])
    is_q4 = 1 if month in [10, 11, 12] else 0
    
    row = {
        'month': month,
        'month_number': month_num,
        'lag_1': lag_1,
        'lag_2': lag_2,
        'lag_3': lag_3,
        'rolling_3': rolling_3,
        'is_q4': is_q4
    }
    
    # Predict and store for next iteration
    pred = model_full.predict(pd.DataFrame([row]))[0]
    forecast_revenue.append(pred)
    
    forecast_months.append({
        'month': date.strftime('%Y-%m'),
        'predicted_revenue': round(pred, 2),
        'is_q4': is_q4
    })

forecast_df = pd.DataFrame(forecast_months)
print("\nNext 6 Months Revenue Forecast:")
print(forecast_df.to_string(index=False))
print(f"\nTotal Predicted Revenue (next 6 months): £{forecast_df['predicted_revenue'].sum():,.0f}")
```
